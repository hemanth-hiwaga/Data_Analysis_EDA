```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
data = pd.read_csv('D:/data_analysis_projects/python_projects/sales_data.csv')

#getting familarized with data
print(data.info())
print(data.describe())
print(data.columns)

#cleaning the data
# removing the 1st column *'Unnamed: 0'* as it is not useful
data = data.drop(columns= 'Unnamed: 0', axis=1)
print(data.columns)
print(data.City.head().tolist)
#observed there is space before city name fixing it
data['City'] = data['City'].str.strip()
print(f'fixed city name: \n{data['City'].head()}')
#checking for nulls and droping them
print(data.isnull().sum())
print('***there are no nulls****')
#now checking for duplicated and dropping them in current dataset
print(f'number of duplicated rows before dropping are: {data.duplicated().sum()}')
#droping duplicates
data.drop_duplicates(inplace= True)
print(f'number of duplicated rows after droppingare: {data.duplicated().sum()}')


#extracting columns for further analysis year, month name, weekday, and total_sales
# 1 formating the the order date
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['year'] = data['Order Date'].dt.year
data['Month_Name'] = data['Order Date'].dt.strftime('%B')
data['week'] = data['Order Date'].dt.strftime('%A')
data['Total_Sales'] = data['Quantity Ordered'] * data['Price Each']
```
